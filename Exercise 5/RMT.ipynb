```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('hhist_2000_75_40.dat',header=None,comment='#')
X=df.drop([0,1],axis=1)
X.columns = range(X.shape[1])
width = df[0]
width=np.mean(width) #actually all equal
x=np.sum(X,axis=0)
hgt=(x/np.sum(x))/width
thr=np.arange(X.shape[1])*width
plt.bar(x=thr,width=width,align='edge',height=hgt)
```

```python
plt.plot(thr, hgt)
```

```python
np.savetxt('rmt_pdf.txt', np.concatenate((thr.reshape(-1,1), hgt.values.reshape(-1,1)), axis=1))
```
